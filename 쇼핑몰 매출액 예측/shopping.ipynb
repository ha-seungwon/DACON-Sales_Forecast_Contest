```python
# date ì—ì„œ ì—°ë„ ì¹¼ëŸ¼, ì›” ì¹¼ëŸ¼, ì¼ ì¹¼ëŸ¼ ë”°ë¡œ ë¶„ë¦¬ 
def seperate_date(df):
    date=list(df['Date'])
    year=[]
    month=[]
    day=[]
    year_=''
    month_=''
    day_=''
    for j in date:
        list_=j.split("/")
        for i in range(3):
            if(i==0):
                day_+=list_[i]
            elif(i==1):
                month_+=list_[i]
            elif(i==2):
                year_+=list_[i]
        year.append(int(year_))
        month.append(int(month_))
        day.append(int(day_))
        year_=''
        month_=''
        day_=''
    df=df.drop(columns='Date')
    df['year']=year
    df['month']=month
    df['day']=day
    return df
```

```python

train=seperate_date(train)
train = train.fillna(train.mean())
train["Weekly_Sales"] = np.log1p(train["Weekly_Sales"])
train["IsHoliday"] = train["IsHoliday"].astype(int)
train = pd.get_dummies(data = train,columns = ['Store'])
train=train.drop(columns=['id'])
```
